```python
text= "The horse ran that way. Chase after it! FAST!"
#all text is the same
lowered_text=text.lower()
clean_text= ""
for character in lowered_text:
  if character.isalpha() or character==" ":
    clean_text=clean_text + character
#count words
words=clean_text.split()
tally={}
for word in words:
    if word in tally:
        tally[word]= tally[word] +1
    else:
        tally[word]=1
        #print count
print (tally)
```
