```python
#start point
x = 3

#step distance
learning_rate = 0.1

#10 steps to bottom
for step in range(10):
    #derivative of x^2
    slope = 2 * x

    #x moves in the opposite direction of slope
    x = x - (slope * learning_rate)

    print(f"Step {step+1}: x is now {x:.4f}, Slope is {slope:.4f}")
```
