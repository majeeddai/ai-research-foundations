```python

import math

#surprisal function
def get_surprisal(probability):
    # Surprisal = -log2 of the probability
    result = -math.log2(probability)
    return result

# Test 1: A very common word (50% chance)
common_word_prob = 0.5
print("Common word surprisal:", get_surprisal(common_word_prob))

# Test 2: A rare word (1% chance)
rare_word_prob = 0.01
print("Rare word surprisal:", get_surprisal(rare_word_prob))

#choose a word and its probability(purely an example here)
word = "the"
probability = 0.12

#function call and save score
score = get_surprisal(probability)

#print info
print(f"The word '{word}' has a surprisal of {score:.2f} bits.")
```
